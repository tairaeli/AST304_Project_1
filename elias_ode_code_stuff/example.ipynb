```python
def DeriveDragModel1D(y, vy):
    '''DerviceDrageModel1D returns the value of the
    position and velocity changes at a given point in 
    phase space for a 1D drag model.
    
    Returns dy_dt and dvy_dt in that order'''
    
    ## Compute the changes; the notation is derviative of thing
    ## underscore with respect to variable, so dvy_dt means
    ## derivative of vy with respect to t
    
    ## Setup parameters (feel free to adjust)
    m = 1
    g = 9.8
    beta = 0.01
    alpha = 0.02
    
    dvy_dt = -g + beta/m*vy + alpha/m*vy**2
    dy_dt = vy
    
    ## Returns position change/dt then velocity change/dt
    return dy_dt, dvy_dt
```

```python
# Initializing base settings for analysis
t0 = 0
tf = 5
N = 10
deltat = (tf-t0)/N
ndim = 1

# Initializing starting conditions
y0 = 0
vy0 = 0

## Run the code. All methods initialize the same way
euler_object = euler(ndim, tf, N, y0, vy0)
rk2_object = rk2(ndim, tf, N, y0, vy0)
rk4_object = rk4(ndim, tf, N, y0, vy0)

y_arr_EC, vy_arr_EC,t_arr_EC = euler_object.iterate(DeriveDragModel1D)
y_arr_RK2, vy_arr_RK2,t_arr_RK2 = rk2_object.iterate(DeriveDragModel1D)
y_arr_RK4, vy_arr_RK4, t_arr_RK4 = rk4_object.iterate(DeriveDragModel1D)
```
